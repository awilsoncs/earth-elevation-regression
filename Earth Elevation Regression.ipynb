# Data

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

plt.rcParams['figure.figsize'] = (10.0, 5.0)

img=mpimg.imread('earth.jpg')
lum_img = img[:,:,0]
plt.imshow(lum_img)
plt.colorbar()
```

# Elevation Plot

Here, we have a graph of the histogram as a function of percentile.

```python
x = [float(x)/200.0 for x in range(0, 200)]
y = [np.percentile(lum_img, i*100) for i in x]
```

```python
plt.plot(x, y)
plt.xlabel('Percentile')
plt.ylabel('Elevation')
plt.show()
```

# Regression

Here, I use a polynomial regression to fit the example set. At n=7, the regression is somewhat underfit, but this suits my purposes just fine. The red dashed line roughly indicates sea level.

```python
n = 7

theta = np.polyfit(x, y, n)
```

```python
def h(x, theta):
    value = [x**n for n in range(n, -1, -1)]
    return np.dot(theta, value)

test = [h(float(x)/200.0, theta) for x in range(0, 200)]
```

```python
plt.plot(x, test, 'b', x, [140 for _ in x], 'r--')
plt.xlabel('Percentile (0.5%)')
plt.ylabel('Elevation')
plt.show()
```
